# **Guiding in Prompts**

```
from dotenv import load_dotenv
import os

load_dotenv()

if os.environ.get('ANTHROPIC_API_KEY'):
    print('Bro Api Key Exists!!!')
else:
    raise ValueError('ANTHROPIC_API_KEY not found')
```

```
from  langchain.chat_models import init_chat_model

llm_message = init_chat_model(model="claude-sonnet-4-6")
```

```
from langchain_core.prompts import PromptTemplate

llm_message.invoke('Tell me a joke. Generate the output in key -value format, setup and punchlines').content

```

# **Using Pydantic Models**

```
from  pydantic import BaseModel, Field

class Llm_schema(BaseModel):
    setup: str = Field(description='The setup for the joke')
    punchline: str = Field(description='The punchline for the joke')
```

```
llm_structured_output = llm_message.with_structured_output(Llm_schema)

result = llm_structured_output.invoke('Tell me a joke')
type(result)
result.punchline
```

# **Typedict**

```
from typing import TypedDict

class llm_schema_td(TypedDict):
    setup: str
    punchline: str
```

```
llm_structured_output_typedict = llm_message.with_structured_output(llm_schema_td)
result = llm_structured_output_typedict.invoke('Tell me a joke')
result['setup']
```
